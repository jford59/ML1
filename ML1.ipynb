```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/drive')
file_path = '/content/drive/My Drive/D3.csv'
sample = pd.DataFrame(pd.read_csv(file_path))
sample.head()

X1 = sample.values[:, 0]
X2 = sample.values[:, 1]
X3 = sample.values[:, 2]
y = sample.values[:, 3]
m = len(y)
n = len(X1)

X_0 = np.ones((m, 1))
X_1 = X1.reshape(m, 1)
X_2 = X2.reshape(m, 1)
X_3 = X3.reshape(m, 1)

x1 = np.hstack((X_0, X_1))
x2 = np.hstack((X_0, X_2))
x3 = np.hstack((X_0, X_3))
x4 = np.hstack((X_0, X_1, X_2, X_3))

theta1 = np.zeros(2)
theta2 = np.zeros(4)

def compute_pred(X, y, theta1):
    predictions = X.dot(theta1)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J

def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_pred(X, y, theta)
    return theta, cost_history

theta1 = [0., 0.]
theta2 = [0., 0., 0., 0.]

iterations = 1500
alpha = 0.02

theta1, cost_history = gradient_descent(x1, y, theta1, alpha, iterations)
f1 = plt.figure()
plt.scatter(x1[:, 1], y, color='blue', marker='+', label='Training Data')
plt.plot(x1[:, 1], x1.dot(theta1), color='green', label='Linear Regression')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Linear Regression Fit')
plt.legend()
plt.show()

f4 = plt.figure()
plt.plot(range(1, iterations + 1), cost_history, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Convergence of gradient descent')
plt.show()

theta1, cost_history = gradient_descent(x2, y, theta1, alpha, iterations)
f2 = plt.figure()
plt.scatter(x2[:, 1], y, color='blue', marker='+', label='Training Data')
plt.plot(x2[:, 1], x2.dot(theta1), color='green', label='Linear Regression')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Linear Regression Fit')
plt.legend()
plt.show()

f5 = plt.figure()
plt.plot(range(1, iterations + 1), cost_history, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Convergence of gradient descent')
plt.show()

theta1, cost_history = gradient_descent(x3, y, theta1, alpha, iterations)
f3 = plt.figure()
plt.scatter(x3[:, 1], y, color='blue', marker='+', label='Training Data')
plt.plot(x3[:, 1], x3.dot(theta1), color='green', label='Linear Regression')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Linear Regression Fit')
plt.legend()
plt.show()

f6 = plt.figure()
plt.plot(range(1, iterations + 1), cost_history, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Convergence of gradient descent')
plt.show()

theta2, cost_history = gradient_descent(x4, y, theta2, alpha, iterations)
theta3 = np.hstack((theta2[0], theta2[1]))
theta4 = np.hstack((theta2[0], theta2[2]))
theta5 = np.hstack((theta2[0], theta2[3]))
f7 = plt.figure()
plt.scatter(x1[:, 1], y, color='blue', marker='+', label='Training Data X1')
plt.scatter(x2[:, 1], y, color='red', marker='+', label='Training Data X2')
plt.scatter(x3[:, 1], y, color='black', marker='+', label='Training Data X3')
plt.plot(x1[:, 1], x1.dot(theta3), color='blue')
plt.plot(x2[:, 1], x2.dot(theta4), color='red')
plt.plot(x3[:, 1], x3.dot(theta5), color='black')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Linear Regression Fit')
plt.legend()
plt.show()

f8 = plt.figure()
plt.plot(range(1, iterations + 1), cost_history, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.title('Convergence of gradient descent')
plt.show()

x4 = np.hstack((1, 1, 1, 1))
y1 = x4.dot(theta2)
print(y1)
x4 = np.hstack((1, 2, 0, 4))
y2 = x4.dot(theta2)
print(y2)
x4 = np.hstack((1, 3, 2, 1))
y3 = x4.dot(theta2)
print(y3)
```
